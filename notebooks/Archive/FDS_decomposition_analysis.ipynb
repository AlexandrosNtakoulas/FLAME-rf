```python
from __future__ import annotations

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.inspection import permutation_importance

from flamekit.io_fronts import Case, load_fronts

# ============================================================
# USER SETTINGS
# ============================================================
PHI = 0.40
LAT_SIZE = "100"
ISOLEVELS = [4.5]

# ---- NEW: aggregate multiple time steps to define curvature thresholds ----
TIME_STEPS = [140, 200, 205, 211]  # <-- put all timesteps you want here

BASE_DIR = Path("../data/isocontours")
POST = True

# Optional Case fields (only used if your Case supports them)
MULTIPLE_RUNS = False
N_RUN = 2
SORET_TERM = False
SORET_POSITIVE = False

# ---- Curvature-based predetermined clusters ----
CURVATURE_COL = "curvature"
# Split aggregated curvature distribution into 3 equal-count parts (tertiles):
Q_LOW = 0.25
Q_HIGH = 0.75

# ------------------------------------------------------------
# Regression / permutation importance settings (independent)
# ------------------------------------------------------------
REG_FEATURES = ["FDS_src_term", "FDS_diffusion_term", "FDS_diff_velocity_term"]
TARGET = "DW_FDS"

# --- Plot controls ---
POINT_SIZE = 10
MARKER = "s"
FIG_DPI = 200

# --- Data thresholds ---
MIN_VALID_SAMPLES_ISO = 200          # skip iso if too few points with finite curvature+x+y
MIN_SAMPLES_PER_CLUSTER_REG = 200    # skip cluster if too few regression samples
TEST_SIZE = 0.25
RANDOM_STATE = 0

# --- Regressor settings ---
RF_N_ESTIMATORS = 400
RF_MAX_DEPTH = None
RF_MIN_SAMPLES_LEAF = 5
RF_N_JOBS = -1

# --- Permutation importance ---
PERM_N_REPEATS = 15

# --- Figure saving ---
SAVE_FIGS = True
OUT_BASE_DIR = Path(r"C:\Users\alexp\Documents\Bachelor Thesis\report_figures\results\FDS_decomposition_analysis")

# Names remain constant in every folder:
XY_FNAME = "xy_plot.png"
CLUSTER_FNAME_FMT = "cluster{cluster}_comp.png"

# ============================================================
# MATPLOTLIB DEFAULTS (better quality)
# ============================================================
plt.rcParams.update({
    "figure.facecolor": "white",
    "savefig.facecolor": "white",
    "axes.grid": False,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.titlesize": 12,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})


# ============================================================
# HELPERS
# ============================================================
def make_case(time_step: int) -> Case:
    """
    Create a Case for a given time_step, using optional fields if supported.
    """
    # Minimal required fields
    base_kwargs = dict(
        base_dir=BASE_DIR,
        phi=PHI,
        lat_size=LAT_SIZE,
        time_step=time_step,
        post=POST,
    )

    # Try extended Case signature (optional fields)
    try:
        return Case(
            **base_kwargs,
            multiple_runs=MULTIPLE_RUNS,
            n_run=N_RUN,
            soret_term=SORET_TERM,
            soret_positive=SORET_POSITIVE,
        )
    except TypeError:
        return Case(**base_kwargs)


def make_out_dir(lat_size: str, time_step: int, c_val: float) -> Path:
    """
    Folder naming convention:
      h_<LAT_SIZE>_t_<TIME_STEP>_T_<ISOLEVEL>
    Example:
      h_100_t_211_T_4.5
    """
    return OUT_BASE_DIR / f"h_{lat_size}_t_{time_step}_T_{c_val:g}"


def require_columns(df: pd.DataFrame, cols: list[str], context: str = "") -> None:
    missing = [c for c in cols if c not in df.columns]
    if missing:
        raise ValueError(f"{context} Missing columns: {missing}. Example available: {list(df.columns[:30])}")


def clean_valid_rows(df: pd.DataFrame, cols_needed: list[str]) -> pd.DataFrame:
    """
    Keep only rows where all cols_needed are finite.
    """
    out = df[cols_needed].replace([np.inf, -np.inf], np.nan).dropna(axis=0, how="any")
    return df.loc[out.index].copy()


def discrete_cluster_cmap(n_clusters: int):
    base = plt.get_cmap("tab10" if n_clusters <= 10 else "tab20")
    colors = [base(i) for i in range(n_clusters)]
    cmap = ListedColormap(colors, name=f"clusters_{n_clusters}")
    boundaries = np.arange(-0.5, n_clusters + 0.5, 1.0)
    norm = BoundaryNorm(boundaries, cmap.N)
    return cmap, norm


def assign_curvature_clusters(curv: np.ndarray, q1: float, q2: float) -> np.ndarray:
    """
    Cluster labels:
      0: most negative curvature   (curv <= q1)
      1: mid curvature             (q1 < curv < q2)
      2: most positive curvature   (curv >= q2)
    """
    # np.digitize with bins [q1, q2] gives 0,1,2 by default for right=False
    #   curv < q1 -> 0
    #   q1 <= curv < q2 -> 1
    #   curv >= q2 -> 2
    return np.digitize(curv, bins=[q1, q2], right=False).astype(int)


def plot_xy_clusters(front_xy: pd.DataFrame, c_val: float, time_step: int, n_clusters: int,
                     q1: float, q2: float, savepath: Path | None = None) -> None:
    cmap, norm = discrete_cluster_cmap(n_clusters)

    fig, ax = plt.subplots(figsize=(6.8, 6.2), dpi=FIG_DPI, constrained_layout=True)
    sc = ax.scatter(
        front_xy["x"], front_xy["y"],
        c=front_xy["cluster"],
        cmap=cmap,
        norm=norm,
        s=POINT_SIZE,
        marker=MARKER,
        linewidths=0,
        alpha=0.95,
        rasterized=True,
    )

    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title(
        f"Curvature-threshold clusters — T={float(c_val):.2f}, t={time_step}\n"
        f"q1={q1:.4e}, q2={q2:.4e}"
    )
    ax.set_aspect("equal", adjustable="box")

    cbar = fig.colorbar(sc, ax=ax, fraction=0.046, pad=0.02)
    cbar.set_label("Cluster ID")
    cbar.set_ticks(np.arange(n_clusters))
    cbar.set_ticklabels([
        "0 (neg)",
        "1 (mid)",
        "2 (pos)",
    ])

    if savepath is not None:
        savepath.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(savepath, dpi=FIG_DPI, bbox_inches="tight")

    plt.show()
    plt.close(fig)


def plot_importances(importances: pd.Series, title: str, savepath: Path | None = None) -> None:
    imp = importances.sort_values(ascending=True)

    fig, ax = plt.subplots(figsize=(8.2, 4.0), dpi=FIG_DPI, constrained_layout=True)
    ax.barh(imp.index, imp.values)
    ax.set_xlabel("Permutation importance (Δ score)")
    ax.set_title(title)
    ax.grid(True, axis="x", linestyle=":", linewidth=0.8, alpha=0.7)

    if savepath is not None:
        savepath.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(savepath, dpi=FIG_DPI, bbox_inches="tight")

    plt.show()
    plt.close(fig)


# ============================================================
# STEP 1: Aggregate multiple time steps to compute curvature thresholds
# ============================================================
print("Aggregating curvature over time steps to compute thresholds...")
curv_all = []

for t in TIME_STEPS:
    case_t = make_case(time_step=t)
    fronts_t = load_fronts(case_t, ISOLEVELS)

    for c_val, front in fronts_t.items():
        require_columns(front, [CURVATURE_COL], context=f"t={t}, ISO {c_val}:")
        curv = (
            front[[CURVATURE_COL]]
            .replace([np.inf, -np.inf], np.nan)
            .dropna(axis=0, how="any")[CURVATURE_COL]
            .to_numpy()
        )
        if curv.size:
            curv_all.append(curv)

if not curv_all:
    raise RuntimeError("No valid curvature samples found across the selected TIME_STEPS.")

curv_all = np.concatenate(curv_all, axis=0)
q1 = float(np.nanquantile(curv_all, Q_LOW))
q2 = float(np.nanquantile(curv_all, Q_HIGH))

print(f"Curvature thresholds computed from aggregated data:")
print(f"  q1 (at {Q_LOW:.3f})  = {q1:.6e}")
print(f"  q2 (at {Q_HIGH:.3f}) = {q2:.6e}")
print(f"  total samples used   = {curv_all.size}")

# ============================================================
# STEP 2: For each time step, apply predetermined thresholds and do plots + saving
# ============================================================
print("\nProcessing each time step individually with fixed thresholds...")

clustered_fronts: dict[tuple[int, float], pd.DataFrame] = {}
summary_rows: list[dict] = []

N_CLUSTERS = 3  # fixed by construction

for t in TIME_STEPS:
    case_t = make_case(time_step=t)
    fronts_t = load_fronts(case_t, ISOLEVELS)

    for c_val, front in fronts_t.items():
        front = front.copy()

        # Required for XY plot + curvature clustering
        require_columns(front, ["x", "y", CURVATURE_COL], context=f"t={t}, ISO {c_val}:")

        # Required for regression later
        require_columns(front, REG_FEATURES + [TARGET], context=f"t={t}, ISO {c_val}:")

        # --- Data for clustering/XY: x,y,curvature must be finite
        front_cluster = clean_valid_rows(front, ["x", "y", CURVATURE_COL])
        if len(front_cluster) < MIN_VALID_SAMPLES_ISO:
            print(f"t={t}, ISO {c_val}: skipped (cluster-valid n={len(front_cluster)} < {MIN_VALID_SAMPLES_ISO})")
            continue

        # Assign predetermined clusters using global thresholds
        front_cluster["cluster"] = assign_curvature_clusters(front_cluster[CURVATURE_COL].to_numpy(), q1, q2)

        # Merge labels back (rows not cluster-valid remain -1)
        front["cluster"] = -1
        front.loc[front_cluster.index, "cluster"] = front_cluster["cluster"].values
        clustered_fronts[(t, float(c_val))] = front

        # Output directory for this time step + isolevel
        out_dir = make_out_dir(LAT_SIZE, t, float(c_val))
        if SAVE_FIGS:
            out_dir.mkdir(parents=True, exist_ok=True)

        # XY plot (cluster-valid rows only)
        xy_save = (out_dir / XY_FNAME) if SAVE_FIGS else None
        plot_xy_clusters(
            front_cluster[["x", "y", "cluster"]],
            c_val=float(c_val),
            time_step=t,
            n_clusters=N_CLUSTERS,
            q1=q1,
            q2=q2,
            savepath=xy_save,
        )

        # --- Train one regressor per cluster (using REG_FEATURES, not curvature)
        for k in [0, 1, 2]:
            dfk_all = front[front["cluster"] == k].copy()
            dfk = clean_valid_rows(dfk_all, REG_FEATURES + [TARGET])
            n_k = len(dfk)

            if n_k < MIN_SAMPLES_PER_CLUSTER_REG:
                print(f"t={t}, ISO {c_val} | Cluster {k}: skipped (reg-valid n={n_k} < {MIN_SAMPLES_PER_CLUSTER_REG})")
                continue

            X = dfk[REG_FEATURES].to_numpy()
            y = dfk[TARGET].to_numpy()

            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
            )

            model = RandomForestRegressor(
                n_estimators=RF_N_ESTIMATORS,
                max_depth=RF_MAX_DEPTH,
                min_samples_leaf=RF_MIN_SAMPLES_LEAF,
                n_jobs=RF_N_JOBS,
                random_state=RANDOM_STATE,
            )
            model.fit(X_train, y_train)

            y_pred = model.predict(X_test)
            r2 = r2_score(y_test, y_pred)
            rmse = float(np.sqrt(mean_squared_error(y_test, y_pred)))

            perm = permutation_importance(
                model, X_test, y_test,
                n_repeats=PERM_N_REPEATS,
                random_state=RANDOM_STATE,
                n_jobs=RF_N_JOBS,
            )
            imp_mean = pd.Series(perm.importances_mean, index=REG_FEATURES)
            imp_std = pd.Series(perm.importances_std, index=REG_FEATURES)

            # Save summary
            row = {
                "time_step": int(t),
                "T": float(c_val),
                "cluster": int(k),
                "n_cluster_reg_valid": int(n_k),
                "r2_test": float(r2),
                "rmse_test": float(rmse),
                "q1": q1,
                "q2": q2,
                "reg_features": ",".join(REG_FEATURES),
                "top_feature": imp_mean.idxmax(),
            }
            for f in REG_FEATURES:
                row[f"perm_imp_mean__{f}"] = float(imp_mean[f])
                row[f"perm_imp_std__{f}"] = float(imp_std[f])
            summary_rows.append(row)

            print("\n" + "-" * 90)
            print(f"t={t} | T={float(c_val):.2f} | Cluster {k} | reg-valid n={n_k}")
            print(f"Curvature thresholds: q1={q1:.6e}, q2={q2:.6e}")
            print(f"Regression features: {REG_FEATURES}")
            print(f"Test R² = {r2:.4f} | Test RMSE = {rmse:.6e}")
            print("Permutation importance (mean ± std):")
            for f in imp_mean.sort_values(ascending=False).index:
                print(f"  - {f}: {imp_mean[f]:.6e} ± {imp_std[f]:.6e}")
            print(f"Top feature: {imp_mean.idxmax()}")

            imp_save = (out_dir / CLUSTER_FNAME_FMT.format(cluster=int(k))) if SAVE_FIGS else None
            plot_importances(
                imp_mean,
                title=f"t={t}, T={float(c_val):.2f} — Cluster {k} — Feature importance for predicting {TARGET}",
                savepath=imp_save,
            )

# ============================================================
# SUMMARY TABLE
# ============================================================
if summary_rows:
    summary = pd.DataFrame(summary_rows).sort_values(["time_step", "T", "cluster"]).reset_index(drop=True)
    print("\n" + "=" * 90)
    print("Summary (per time step, per isolevel, per cluster)")
    print("=" * 90)
    with pd.option_context("display.max_columns", 200, "display.width", 220):
        print(summary)
else:
    print("No cluster models were trained (insufficient valid samples or clusters too small).")
```
